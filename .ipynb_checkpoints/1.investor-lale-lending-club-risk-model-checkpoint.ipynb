```python
credentials_1 = {
    'username': 'username',
    'password': """password""",
    'sg_service_url': 'url',
    'database': 'BLUDB',
    'host': 'host',
    'port': 'port',
    'auto_discovery': 'true',
}

df = IdaDataFrame(idadb_c568b279819b490d915961aba3c756c5, 'BXJ33585.LENDINGCLUB').as_dataframe()
df.columns = df.columns.str.lower()
```

```python
columns = ['id','application_type','loan_amnt','purpose','annual_inc','fico_range_low','term','dti','revol_bal','revol_util',
           'inq_last_6mths','delinq_2yrs','pub_rec','total_bal_il','total_rev_hi_lim','avg_cur_bal','tot_hi_cred_lim', 
           'grade']
```

```python
df2 = df[columns].copy()
```

```python
pd.set_option('display.max_rows', 400)
```

## Drop Null Values

```python
df2 = df2.dropna()
df2['grade'].value_counts()
```

## Resample

```python
X, y = df2.drop('grade', axis=1), df2['grade']
```

```python
from imblearn.under_sampling import RandomUnderSampler
cc = RandomUnderSampler(random_state=0)
X_resampled, y_resampled = cc.fit_resample(X, y)
```

```python
y_resampled.value_counts()
```

```python
df[df['id'].isin(X_resampled['id'].unique())].to_csv('../datasets/lending-club-loan-v2/data-all.csv', index=False)
```

```python
def transform_purpose(value):
    if value in ['debt_consolidation', 'credit_card','medical', 'house', 'car', 'small_business', 'other']:
        return value
    else:
        return 'personal'

X_resampled['purpose'] = X_resampled['purpose'].apply(transform_purpose)
```

```python
def strip(value):
    return str(value).strip()
```

```python
for i in X_resampled.select_dtypes('object').columns:
    X_resampled[i] = X_resampled[i].apply(strip) 
```

## Split Training Testing Set

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y_resampled = le.fit_transform(y_resampled)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_resampled.drop('id', axis=1), y_resampled, stratify=y_resampled, random_state=0)
```

```python
X_train_reindex = X_train.reset_index().drop('index', axis=1)
X_test_reindex = X_test.reset_index().drop('index', axis=1)
```

## Modeling

```python
from lale.lib.lale import Project, NoOp
from lale.lib.lale import ConcatFeatures as Concat
from sklearn.preprocessing import OneHotEncoder as OneHot
from sklearn.preprocessing import Normalizer as Norm
from sklearn.decomposition import PCA
from xgboost import XGBClassifier as xgboost
from sklearn.ensemble import RandomForestClassifier as RF
from sklearn.tree import DecisionTreeClassifier as Tree
from sklearn.svm import SVC
from sklearn.feature_selection import RFE
import lale

lale.wrap_imported_operators()
```

```python
auto_planned = (
    (Project(columns={'type':'number'}) >> (Norm | NoOp) &
     Project(columns={'type':'string'}) >> OneHot)  
    >> Concat
    >> (RF | xgboost | SVC | Tree)
)

from lale.lib.lale.hyperopt import Hyperopt
auto_optimizer = Hyperopt(estimator=auto_planned, cv=3, max_evals=10)

auto_trained = auto_optimizer.fit(X_train, y_train)

# y_proba_test = auto_trained.predict_proba(pool_test)[:, 1]
y_predict = auto_trained.predict(X_test)
p_test = precision_score(y_test, y_predict,average='macro')
r_test = recall_score(y_test, y_predict,average='macro')
accuracy = accuracy_score(y_test, y_predict)
print(f'''Precision (test): {p_test:.3f}
Recall (test): {r_test:.3f}
Accuracy (test): {accuracy:.3f}''')

cm_test = confusion_matrix(le.inverse_transform(y_test), le.inverse_transform(y_predict))
ax = sns.heatmap(cm_test, cmap='viridis_r', annot=True, fmt='d', square=True, xticklabels=le.classes_, yticklabels=le.classes_)
ax.set_xlabel('Predicted')
ax.set_ylabel('True');

auto_planned.visualize()
```

### Using CatBoost to do feature ranking

```python
from catboost import CatBoostClassifier, Pool
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, le.transform(y), stratify=le.transform(y), random_state=0)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, stratify=y_train, random_state=0)
```

```python
cat_feat_ind = (X_train.dtypes == 'object').nonzero()[0]
pool_train = Pool(X_train, y_train, cat_features=cat_feat_ind)
pool_val = Pool(X_val, y_val, cat_features=cat_feat_ind)
pool_test = Pool(X_test, y_test, cat_features=cat_feat_ind)

model = CatBoostClassifier(learning_rate=0.03,
                           iterations=1000,
                           early_stopping_rounds=100,
                            loss_function='MultiClass',
                           verbose=False,
                           random_state=0)
model.fit(pool_train, eval_set=pool_val, plot=True);
```

```python
feat = model.feature_names_
imp = model.feature_importances_
df = pd.DataFrame({'Feature': feat, 'Importance': imp})
df = df.sort_values('Importance', ascending=False)[:10]
sns.barplot(x='Importance', y='Feature', data=df);
```

```python
df.to_csv('../datasets/lending-club-loan-v2/top-10-features.csv', index=False)
```

```python
y_testPredict = auto_trained.predict(X_test)
y_trainPredict = auto_trained.predict(X_train)

X_trainSave = X_train.copy()
X_trainSave['grade_predicted'] = le.inverse_transform(y_trainPredict)

X_testSave = X_test.copy()
X_testSave['grade_predicted'] = le.inverse_transform(y_testPredict)

X_Save = X_trainSave.append(X_testSave)

df_resampled = X_resampled.copy()
df_resampled['grade'] = le.inverse_transform(y_resampled)

df_resampled = pd.concat([df_resampled, X_Save[['grade_predicted']].sort_index()], axis=1)

df_all = pd.merge(df, df_resampled[['id','grade_predicted']], on='id')

df_2A = df_all[(df_all['grade'] == 'A') & (df_all['grade_predicted'] == 'A')].head(2)

df_3B = df_all[(df_all['grade'] == 'B') & (df_all['grade_predicted'] == 'B')].head(3)

df_4C = df_all[(df_all['grade'] == 'C') & (df_all['grade_predicted'] == 'C')].head(4)

df_1D = df_all[(df_all['grade'] == 'D') & (df_all['grade_predicted'] == 'D')].head(1)

df_10 = df_2A.append(df_3B).append(df_4C).append(df_1D)

df_all.to_csv('../datasets/lending-club-loan-v2/data-all.csv', index=False)

df_10.to_csv('../datasets/lending-club-loan-v2/data-10-selected.csv', index=False)
```
